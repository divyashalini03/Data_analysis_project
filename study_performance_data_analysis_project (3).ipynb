```python
import pandas as pd
import matplotlib.pyplot as plt
# Histogram for math scores
plt.figure(figsize=(8, 5))
ds['math_score'].plot(kind='hist', bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Box plot for math scores
plt.figure(figsize=(8, 5))
ds['math_score'].plot(kind='box', vert=False, patch_artist=True, color='lightblue')
plt.title('Box Plot of Math Scores')
plt.xlabel('Math Score')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'ds' is your DataFrame with the provided columns

# Define categorical columns
categorical_columns = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']

# Plot pie charts for each categorical variable
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    ds[column].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.tab10.colors)
    plt.title(f'Distribution of {column}')
    plt.ylabel('')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'ds' is your DataFrame with the provided columns

# Group the data by parental level of education and calculate the mean scores
mean_scores_by_education = ds.groupby('parental_level_of_education')[['math_score', 'reading_score', 'writing_score']].mean()

# Plot line chart for mean scores
mean_scores_by_education.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Mean Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Subject')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'ds' is your DataFrame with the provided columns

# Define categorical columns
categorical_columns = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']

# Plot bar charts for each categorical variable
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    ds[column].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
